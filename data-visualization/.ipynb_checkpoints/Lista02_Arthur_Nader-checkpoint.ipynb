### 1) BloodPressure01

#### Blood pressure histogram

```python
plt.figure(figsize=(5, 3))
plt.hist(df["sbp"], bins = 9, range = (60, 240), edgecolor = "black")
plt.xlabel("Pressure [,\"sbp\"]", fontsize = 10)
plt.ylabel("Frequency", fontsize = 10)
plt.title("Pressure Histogram [,\"sbp\"]\n", fontsize = 12)
plt.xticks(ticks = [100, 150, 200])
plt.show()
```

#### Other histograms

```python
fig, axs = plt.subplots(2, 2, figsize = (10,8))

axs[0, 0].hist(df["sbp"], bins = 9, range = (60, 240), edgecolor = "black")
axs[0, 0].set_xlabel("sbp\n", fontsize = 10)
axs[0, 0].set_ylabel("frequency", fontsize = 10)
axs[0, 0].set_title("Histogram of sbp\n", fontsize = 12)
axs[0, 0].set_xticks(ticks = [100, 150, 200])

axs[0, 1].hist(df["sbp"], bins = 9, range = (60, 240), density = True, edgecolor = "black")
axs[0, 1].set_xlabel("sbp\n", fontsize = 10)
axs[0, 1].set_ylabel("Density", fontsize = 10)
axs[0, 1].set_title("Histogram of sbp\n", fontsize = 12)
axs[0, 1].set_xticks(ticks = [100, 150, 200])

axs[1, 0].hist(df["sbp"], bins = [50, 130,160,200, max(df["sbp"])], range = (60, 240), density = True, edgecolor = "black")
axs[1, 0].set_xlabel("sbp", fontsize = 10)
axs[1, 0].set_ylabel("Density", fontsize = 10)
axs[1, 0].set_title("\nPS\n", fontsize = 12)
axs[1, 0].set_xticks(ticks = [50, 100, 150, 200])

axs[1, 1].hist(df["sbp"], bins = 15, range = (60, 240), density = True, edgecolor = "black")
axs[1, 1].set_xlabel("pressure", fontsize = 10)
axs[1, 1].set_ylabel("Density", fontsize = 10)
axs[1, 1].set_title("\nSystolic pressure\n", fontsize = 12)
axs[1, 1].set_xticks(ticks = [100, 150, 200])

fig.tight_layout()
plt.show()
```

### 2) POP2006

#### Reading the data

```python
df_pop = pd.read_csv('POP2006.csv')
```

#### Histograms

```python
df_aux1 = df_pop[df_pop["POP2006"] < 1000000]
df_aux2 = df_pop[df_pop["POP2006"] < 500000]
df_aux3 = df_pop[df_pop["POP2006"] < 100000]
```

```python
fig, axs = plt.subplots(2, 2, figsize = (12,8))

axs[0, 0].hist(df_pop["POP2006"], bins = 12, edgecolor = "black")
axs[0, 0].set_title("Municipal population in 2006\n", fontsize = 12)
axs[0, 0].set_xticks(ticks = [0,2e6,4e6,6e6,8e6,10e6,12e6])

axs[0, 1].hist(df_aux1["POP2006"], bins = 10, edgecolor = "black")
axs[0, 1].set_title("pop < 1 million\n", fontsize = 12)
axs[0, 1].set_xticks(ticks = [0,2e5,4e5,6e5,8e5,10e5])

axs[1, 0].hist(df_aux2["POP2006"], bins = 10, edgecolor = "black")
axs[1, 0].set_title("pop < 500 thousand\n", fontsize = 12)
axs[1, 0].set_xticks(ticks =[0,1e5,2e5,3e5,4e5,5e5])

axs[1, 1].hist(df_aux3["POP2006"], bins = 10, edgecolor = "black")
axs[1, 1].set_title("pop < 100 thousand\n", fontsize = 12)
axs[1, 1].set_xticks(ticks = [0,2e4,4e4,6e4,8e4,10e4])

fig.tight_layout()
plt.show()
```

## Logathmic histrograms

```python
df_log = np.log10(df_pop["POP2006"])
```

```python
fig, axs = plt.subplots(1, 2, figsize = (8,3))

axs[0].hist(df_log, bins = 8, edgecolor = "black")
axs[0].set_title("Population, log10 scale\n", fontsize = 12)
axs[0].set_xticks(ticks = [3,4,5,6,7])

axs[1].hist(df_log, bins = 8, edgecolor = "black")
axs[1].set_title("Labels on the original scale\n", fontsize = 12)
axs[1].set_xticks(ticks = [3,4,5,6,7])
axs[1].set_xticklabels(labels = ["10^3","10^4","10^5","10^6","10^7"])

fig.tight_layout()
plt.show()
```

### 4)  BloodPressure01 - Boxplots

#### Box-plot

```python
plt.figure(figsize=(4, 4))
plt.boxplot(df['sbp'])
plt.yticks(ticks = [100, 150, 200])
plt.xticks([])
plt.ylabel("Pressure", fontsize = 12)
plt.title("Boxplot of pressure", fontsize = 14)
plt.show()
```

#### Other boxplots

```python
df_w1 = df[df["overwt"] == 1]
df_w2 = df[df["overwt"] == 2]
df_w3 = df[df["overwt"] == 3]
dados = [df_w1["sbp"], df_w2["sbp"], df_w3["sbp"]]
```

```python
fig = plt.figure(figsize =(8, 4))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(dados)
 
plt.yticks(ticks = [100, 150, 200])
plt.ylabel("Pressure", fontsize = 12)
plt.xlabel("1=Normal, 2=Overweight, 3=Obese", fontsize = 12)
plt.title("Pressure by overweight", fontsize = 14)
plt.show()
```

```python
df_i1 = df[df["income"] == 1]
df_i2 = df[df["income"] == 2]
df_i3 = df[df["income"] == 3]
dados_renda = [df_i1["sbp"], df_i2["sbp"], df_i3["sbp"]]
```

```python
fig = plt.figure(figsize =(8, 4))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(dados_renda)
 
plt.yticks(ticks = [100, 150, 200])
plt.ylabel("Pressure", fontsize = 12)
plt.xlabel("1=Low, 2=Middle, 3=High", fontsize = 12)
plt.title("Pressure by income", fontsize = 14)
plt.show()
```

```python
df_M = df[df["gender"] == "M"]
df_F = df[df["gender"] == "F"]

df_aux1 = df_F[df_F["overwt"] == 1]
df_aux2 = df_F[df_F["overwt"] == 2]
df_aux3 = df_F[df_F["overwt"] == 3]

df_aux4 = df_M[df_M["overwt"] == 1]
df_aux5 = df_M[df_M["overwt"] == 2]
df_aux6 = df_M[df_M["overwt"] == 3]

dados_misturados = [df_aux1["sbp"], df_aux2["sbp"], df_aux3["sbp"], df_aux4["sbp"], df_aux5["sbp"], df_aux6["sbp"]]
```

```python
fig = plt.figure(figsize =(8, 4))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(dados_misturados)
 
plt.yticks(ticks = [100, 150, 200])
plt.ylabel("Pressure", fontsize = 12)
plt.xticks(ticks = [1,2,3,4,5,6], labels = ["1.F", "2.F","3.F","1.M","2.M","3.M"])
plt.xlabel("Overweight and sex", fontsize = 12)
plt.title("Pressure by overweight and sex", fontsize = 14)
plt.show()
```

```python
df_t0 = df[df["trt"] == 0]
df_t1 = df[df["trt"] == 1]

dados_tratamento = [df_t0["sbp"], df_t1["sbp"]]
```

```python
fig = plt.figure(figsize =(6, 4))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(dados_tratamento)
 
plt.yticks(ticks = [100, 150, 200])
plt.xticks(ticks = [1,2], labels = ["0", "1"])
plt.ylabel("Pressure", fontsize = 12)
plt.xlabel("1=Treated, 0=Untreated", fontsize = 12)
plt.title("Pressure by previous treatment for hypertension", fontsize = 14)
plt.show()
```

### 5) Mtcars

#### Reading the data

```python
df_carros = pd.read_csv('mtcars.csv')
```

#### Scatterplot matrix

```python
df_mpg = df_carros["mpg"]
df_disp = df_carros["disp"]
df_drat = df_carros["drat"]
df_wt = df_carros["wt"]

atributos = ["mpg", "disp", "drat", "wt"]
df_atributos = [df_mpg, df_disp, df_drat, df_wt]
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(10,10))
fig.subplots_adjust(hspace=0.10, wspace=0.10)

axes_off = [0, 2, 5, 6, 7, 8, 9, 10, 13, 15]
x_off = [4, 11]
y_off = [1, 14]
subplots_indexes = [(0,1), (0,2), (0,3), (1,2), (1,3), (2,3)]

for ax in range(len(axes.flat)):
    
    if ax in axes_off :
        axes.flat[ax].xaxis.set_visible(False)
        axes.flat[ax].yaxis.set_visible(False)
        
    elif ax in x_off:
        axes.flat[ax].xaxis.set_visible(False)
        
    elif ax in y_off:
        axes.flat[ax].yaxis.set_visible(False)
        
    if ax == 1:
        axes.flat[ax].xaxis.set_ticks_position('top')
        
    elif ax == 3:
        axes.flat[ax].xaxis.set_ticks_position('top')
        axes.flat[ax].yaxis.set_ticks_position('right')

    elif ax == 11:
        axes.flat[ax].yaxis.set_ticks_position('right')

for i, j in subplots_indexes:
    axes[j,i].scatter(df_atributos[i], df_atributos[j])
    axes[i,j].scatter(df_atributos[j], df_atributos[i])

for i in range(len(atributos)):
    axes[i,i].annotate(atributos[i], (0.5, 0.5), xycoords='axes fraction', ha='center', va='center', fontsize = 14)

fig.suptitle('Simple Scatterplot Matrix', fontsize = 20)
plt.show()
```
