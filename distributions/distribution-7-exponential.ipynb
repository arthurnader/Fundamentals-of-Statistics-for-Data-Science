The exponential is a continuous probability distribution that is often used to model the time between events that occur independently and at a constant rate.

The probability density function of the exponential distribution is given by:

$$
f(x; \lambda) = \lambda e^{-\lambda x}, \quad x \ge 0
$$

where:
- Î» is the rate of events per unit of time.
- x is the random variable representing the time between events.


## Exponential distibution

```python
def exp(lambda_, x): ## remember lambda is a keyword in python
    
    return stats.expon.pdf(x, scale = 1/lambda_);
```

```python
x = np.linspace(0, 10, 100)
```

```python
lambda_1 = 0.5
lambda_2 = 1
lambda_3 = 2
```

```python
pdf1 = exp(lambda_1, x)
pdf2 = exp(lambda_2, x)
pdf3 = exp(lambda_3, x)
```

```python
plt.figure(figsize = (10,5))

plt.plot(x, pdf1, color='blue')
plt.plot(x, pdf2, color='green')
plt.plot(x, pdf3, color='red')
plt.title("PDF of exponential for different lambdas", fontsize = 18)
plt.xlabel('Values', fontsize = 14)
plt.ylabel('Probability Density', fontsize = 14)
plt.legend([0.5, 1, 2])
plt.show()
```

Notice how smaller values of lambda âflattenâ the curve increasing the chance of larger values

The expected value is simply 1/lambda. Which makes sense considering the comment above. Large values tend to shift the expected value to the right.

```python
expected = stats.expon(scale=1/lambda_1).mean()
lambda_1, expected
```

## First day at new job

The time between two phone calls in a store is modeled by an exponential distribution and is expected to be 15 minutes:

```python
lambda_15 = 1/15
x = np.linspace(0, 50)
```

```python
pdf = exp(lambda_15, x)
```

```python
plt.figure(figsize = (10,5))

plt.plot(x, pdf, color='blue')
plt.title("PDF for lambda = 1/15", fontsize = 18)
plt.xlabel('Values', fontsize = 14)
plt.ylabel('Probability Density', fontsize = 14)
plt.show()
```

On store promotion days, this value change to 4 minutes:

```python
lambda_4 = 1/4
x = np.linspace(0, 50)
```

```python
pdf = exp(lambda_4, x)
```

```python
plt.figure(figsize = (10,5))

plt.plot(x, pdf, color='blue')
plt.title("PDF for lambda = 0.25", fontsize = 18)
plt.xlabel('Values', fontsize = 14)
plt.ylabel('Probability Density', fontsize = 14)
plt.show()
```

Marco started working there today and doesn't know if it's promotion day or not, but he collected the following data in an hour answering calls:

```python
data = [7,11,8,10,5,14]
```

The values are between the expected values of the two distributions, so he finds it difficult to answer without calculations. What do you think? Is it promotion day?

```python
p_4 = np.product(stats.expon.pdf(data, scale = 1/lambda_4))
p_15 = np.product(stats.expon.pdf(data, scale = 1/lambda_15))
```

After the calculations, Marco says:

```python
if p_4 > p_15:
    print(f"I'm {round(p_4*100/(p_4+p_15),3)}% sure is a promotion day")
else:
    print(f"I'm {round(p_15*100/(p_4+p_15),3)}% sure is a normal day")
```
